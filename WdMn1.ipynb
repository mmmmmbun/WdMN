**Matrix Operations**

Above, we gave you 3 examples: addition of two matrices, multiplication of two matrices and transpose of a matrix. We used nested lists before to write those programs. 
Let's see how we can do the same task using NumPy array.

**Addition of Two Matrices**

We use + operator to add corresponding elements of two NumPy matrice

```python
A = np.array([[2, 4], [5, -6]])
B = np.array([[9, -3], [3, 6]])
C = A + B      # element wise addition
print(C)
```

**Multiplication of Two Matrices**

To multiply two matrices, we use dot() method. Learn more about how numpy.dot works.

Note: * is used for array multiplication (multiplication of corresponding elements of two arrays) not matrix multiplication.

```python
A = np.array([[3, 6, 7], [5, -3, 0]])
B = np.array([[1, 1], [2, 1], [3, -3]])
C = A.dot(B)
print(C)
```

**Transpose of a Matrix**

We use numpy.transpose to compute transpose of a matrix.

```python
A = np.array([[1, 1], [2, 1], [3, -3]])
print(A.transpose())
```

***Inverse Matrix***

```python
A = np.array([[1, 4, 5], 
    [-5, 8, 9],
    [-6, 7, 11]])

np.linalg.inv(A)
```

***Matrix Determinant***

```python
A = np.array([[1, 4, 5], 
    [-5, 8, 9],
    [-6, 7, 11]])

np.linalg.det(A)
```

***Matrix Eigenvalues/Eigenvectors***

```python
A = np.array([[1, 4, 5], 
    [-5, 8, 9],
    [-6, 7, 11]])

w, v = np.linalg.eig(A)
print(w)
print(v)
```

***Access matrix elements, rows and columns***

**Access matrix elements**

Similar like lists, we can access matrix elements using index. Let's start with a one-dimensional NumPy array.

```python
A = np.array([2, 4, 6, 8, 10])

print("A[0] =", A[0])     # First element     
print("A[2] =", A[2])     # Third element 
print("A[-1] =", A[-1])   # Last element     
```

```python
A = np.array([[1, 4, 5, 12],
    [-5, 8, 9, 0],
    [-6, 7, 11, 19]])

#  First element of first row
print("A[0][0] =", A[0][0])  

# Third element of second row
print("A[1][2] =", A[1][2])

# Last element of last row
print("A[-1][-1] =", A[-1][-1])   
```

**Access rows of a Matrix**

```python
A = np.array([[1, 4, 5, 12], 
    [-5, 8, 9, 0],
    [-6, 7, 11, 19]])

print("A[0] =", A[0]) # First Row
print("A[2] =", A[2]) # Third Row
print("A[-1] =", A[-1]) # Last Row (3rd row in this case)
```

**Access columns of a Matrix**

```python
A = np.array([[1, 4, 5, 12], 
    [-5, 8, 9, 0],
    [-6, 7, 11, 19]])

print("A[:,0] =",A[:,0]) # First Column
print("A[:,3] =", A[:,3]) # Fourth Column
print("A[:,-1] =", A[:,-1]) # Last Column (4th column in this case)
```

```python
letters = np.array([1, 3, 5, 7, 9, 7, 5])

# 3rd to 5th elements
print(letters[2:5])        # Output: [5, 7, 9]

# 1st to 4th elements
print(letters[:-5])        # Output: [1, 3]   

# 6th to last elements
print(letters[5:])         # Output:[7, 5]

# 1st to last elements
print(letters[:])          # Output:[1, 3, 5, 7, 9, 7, 5]

# reversing a list
print(letters[::-1])          # Output:[5, 7, 9, 7, 5, 3, 1]
```

```python
A = np.array([[1, 4, 5, 12, 14], 
    [-5, 8, 9, 0, 17],
    [-6, 7, 11, 19, 21]])

print(A[:2, :4])  # two rows, four columns

print(A[:1,])  # first row, all columns

print(A[:,2])  # all rows, second column

print(A[:, 2:5])  # all rows, third to the fifth column
```

```python
def norma1(vector):
    return abs(float(sum(vector)))
```

```python
def norma2(vector):
    if len(vector)==1:
         return abs(float(vector[0]))
    else:
        papa = [x**2 for x in vector]
        return np.sqrt(sum(papa))
```

```python
A=[2,5,7,8,9]
print(norma2(A))
np.linalg.norm(A)
```

```python
def norm4(matrix):
    pocz=0
    if len(matrix)==1:
         return abs(sum(matrix))
    else:
        for element in matrix: 
            papa = [x**2 for x in element]
            pocz +=sum(papa)
    return np.sqrt(pocz)
```

```python
print(norm4(A))
np.linalg.norm(A)
```

```python
def zad10(tablica:list):
    sred = sum(tablica)/len(tablica)
    warian_sum = sum([(x-sred)**2 for x in tablica])
    warian = warian_sum/(len(tablica)-1)
    odch = np.sqrt(warian)
    return [sred, warian, odch]
```

```python
def zad20(a,b,c):
    delta=b**2-4*a*c
    if delta == 0:
        x0 = -b/2*a
        return x0
    elif delta > 0:
        x1=(-b-np.sqrt(delta))/2*a
        x2=(-b+np.sqrt(delta))/2*a
        return (x1, x2)
    else:
        print("Nie ma pierwiastków")
        
```

```python
zad20(1,5,6)
```
